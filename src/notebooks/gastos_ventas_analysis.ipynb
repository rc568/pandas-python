# CORRECCIONES

## Duplicados

1. Duplicate exact rows

```python
filter_duplicates = df_gastos_ventas.duplicated(keep=False)
df_gastos_ventas.iloc[df_gastos_ventas[filter_duplicates].index, 0:12]
```

2. Products with same code but differente name

```python
df = df_gastos_ventas.drop_duplicates(subset=['Código', 'Producto']).sort_values(by='Código', ascending=True)
filter_duplicates = df.duplicated(subset=['Código'], keep=False)
df.loc[filter_duplicates, ['Código', 'Producto']]
```

3. Companies with same RUC but differente name

```python
# pd.set_option('display.max_rows', None)
pd.reset_option('display.max_rows', None)
df_gastos_ventas['Nombre'] = df_gastos_ventas['Nombre'].str.strip()
df = df_gastos_ventas.drop_duplicates(subset=['RUC/DNI', 'Nombre']).sort_values(by='Código', ascending=True)
filter_duplicates = df.duplicated(subset=['RUC/DNI'], keep=False)
df_unique_dni_ruc = df.loc[filter_duplicates, ['RUC/DNI', 'Nombre']].dropna(subset=['RUC/DNI']).astype({'RUC/DNI': 'string'}).sort_values(by='RUC/DNI')
df_unique_dni_ruc
```

4. Invoice code with different company identifier

5. Invoice code with different company identifier

```python
filter = df_gastos_ventas['Observación'].str.startswith('E') & df_gastos_ventas.duplicated(subset=['Observación', 'RUC/DNI'], keep=False)
df = df_gastos_ventas.loc[filter, ['RUC/DNI', 'Observación']].drop_duplicates(subset=['RUC/DNI', 'Observación'])
df[df.duplicated()]
```

6. Get the invoices code and the missing ones

```python
regex_invoice_code = r'E\d{3}-\d{1,4}'
invoices_set = { f'E001-{num}' for num in range(491, 2994) }

filt = df_gastos_ventas['Observación'].str.contains(regex_invoice_code, na=False)
df_unique_invoice_code = df_gastos_ventas.loc[filt, ['Observación']].drop_duplicates()
list_unique_invoice_code_from_gastos = set(df_unique_invoice_code['Observación'].to_list())

difference = list(invoices_set - list_unique_invoice_code_from_gastos)

pd.Series(difference).to_csv('../output/invoices_not_in_gastos.csv', encoding='utf-8')
```

```python
regex_invoice_code = r'E\d{3}-(\d{1,4})'

df_invoices = df_gastos_ventas['Observación'].str.extract(regex_invoice_code).dropna().astype(int)
df_invoices.sort_values(by=0, ascending=True, inplace=True)
df_invoices
```

7. Sale with invoice code but without company identifier or company name

```python
filter_missing_invoices_name = ~df_gastos_ventas['Observación'].isna() & df_gastos_ventas['RUC/DNI'].isna()
regex_boleta = r'EB\d{2}-\d{1,4}'
regex_factura = r'E\d{3}-\d{1,4}'

df = df_gastos_ventas.loc[filter_missing_invoices_name, ['Observación']].rename(columns={'Observación': 'num_comprobante'})
df['num_comprobante'] = df['num_comprobante'].str.replace(" ", "").drop_duplicates().sort_values()

df_boleta = df[df['num_comprobante'].str.contains(regex_boleta, na=False)]
df_factura = df[df['num_comprobante'].str.contains(regex_factura, na=False)]

df_boleta['num_comprobante'].to_csv('../output/invoices_with_missing_company_code_boleta.csv', encoding='utf-8', index=False)
df_factura['num_comprobante'].to_csv('../output/invoices_with_missing_company_code_factura.csv', encoding='utf-8', index=False)
```

8. Invoices with invalid codes (EXXX-XXXX, EBXX-XXXX)

```python
regex = r'(EB\d{2}|E\d{3})-\d{1,4}$'

filt = df_gastos_ventas['Observación'].str.match(regex, na=False)
df_gastos_ventas.loc[~filt, 'Observación'].drop_duplicates()
```

# Export to CSV

1. Apply strip to columns

```python
df = df_gastos_ventas.iloc[:, 0:12]
df['Nombre'] = "'" + df['Nombre'].astype('string').str.strip()
df['RUC/DNI'] = "'" + df['RUC/DNI'].astype('string').str.strip()
df['Número'] = "'" + df['Número'].astype('string').str.strip()
df['Observación'] = "'" + df['Observación'].astype('string').str.strip()
df.to_csv('../output/gastos_ventas_cleaned.csv', index=False, encoding='utf-8-sig')

df.head(15)
```
