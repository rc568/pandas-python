## Sorting


```python
df_lista_without_na = df_lista.drop(index=df_lista[~filt_by_code].index)
df_lista_without_na.drop(
    columns=["r", "Stock Alerta", "Total Venta", "Total Ganancia", "Capital"],
    inplace=True,
)
df_lista_without_na.sort_index(inplace=True)
df_lista_without_na.sort_values(
    by=["Porcentaje", "Precio Venta"], ascending=[False, True]
)
```

## Group by


```python
df_lista["Precio Venta"].median()
```

```python
df_ventas["Medio"].value_counts(normalize=True)
```

```python
sales_channel_group = df_ventas.groupby("Medio")
sales_channel_group["Código"].value_counts().loc["Linio"]
```

```python
sales_channel_group["Total Venta"].agg(["median", "mean", "sum", "min", "max"])
```

```python
total_sales_channel = df_ventas["Medio"].value_counts()
# .sum() se aplica a la Serie de tipo bool
total_sales_channel_by_name = sales_channel_group["Nombre"].apply(
    lambda x: x.str.contains("Das").sum()
)

total_sales_df = pd.concat(
    [total_sales_channel, total_sales_channel_by_name], axis="columns", sort=True
)
total_sales_df.rename(
    columns={"count": "Total per medium", "Nombre": "Total by name"}, inplace=True
)
total_sales_df["Total by name (%)"] = (
    total_sales_df["Total by name"] / total_sales_df["Total per medium"]
) * 100
total_sales_df.sort_values(by="Total by name (%)", ascending=False, inplace=True)

total_sales_df
```
